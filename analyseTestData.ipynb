```python
remove_words = ["@handle", "RT"]

def pre_process(sentence):
    sentence = sentence.split()
    target_remove = set()
    for token in sentence:
        for target in remove_words:
            if target in token:
                target_remove.add(token)
                
    for target in target_remove:
        sentence.remove(target)
    
    sentence = ' '.join(sentence)
    return sentence

def pre_process_zwz(sentence):
    oldSentence = sentence
    sentence = sentence.split()
    target_remove = set()

    if len(sentence) < 3:
        print("--------------------------")
        print("Sentence: {0}".format(oldSentence))
    
    for token in sentence:
        for target in remove_words:
            if (target != token) and (target in token):
                #print("--------------------------")
                #print("Sentence: {0}".format(oldSentence))
                #print("Token: {0}".format(token))
                target_remove.add(token)
    for target in target_remove:
        sentence.remove(target)
    
    sentence = ' '.join(sentence)
    return sentence
```
